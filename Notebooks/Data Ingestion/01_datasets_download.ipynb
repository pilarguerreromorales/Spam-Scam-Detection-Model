## Purpose
Provide instructions to manually download and correctly place the datasets used throughout the project from the data sources.

## Dataset Sources

This project uses the following datasets:  

- **SMS Phishing Dataset for Machine Learning and Pattern Recognition**  
  Source: [Mendeley Data - SMS Phishing Dataset for Machine Learning and Pattern Recognition](https://data.mendeley.com/datasets/f45bkkt8pr/1)  
  Download the file named "Dataset_5971.csv" manually from the website.  
  After downloading, rename the file to `filtered_data.csv` (if necessary) and place the file inside the `data/` folder at the root of the project.  

- **SMS Spam Collection Dataset**  
  Source: [Kaggle - SMS Spam Collection Dataset](https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset?resource=download)  
  Download the dataset manually from Kaggle.  
  After downloading, rename the file to `spam.csv` (if necessary) and place the file inside the `data/` folder at the root of the project.  

- **Telegram Spam or Ham Dataset**  
  Source: [Kaggle - Telegram Spam or Ham](https://www.kaggle.com/datasets/mexwell/telegram-spam-or-ham)  
  Download the dataset manually from Kaggle.  
  After downloading, rename the file to `telegram_spam.csv` (if necessary) and place the file inside the `data/` folder at the root of the project.  

- **Sms Spam Dataset**  
  Source: [Kaggle - Sms Spam Dataset](https://www.kaggle.com/datasets/dilpreetkaur17/sms-spam-dataset)  
  Download the dataset manually from Kaggle.  
  After downloading, rename the file to `sms_spam.csv` (if necessary) and place the file inside the `data/` folder at the root of the project.  

- **Spam-Ham Classifier Dataset**  
  Source: [Kaggle - Spam-Ham Classifier Dataset](https://www.kaggle.com/datasets/ilyazored/spam-ham-classifier-dataset)  
  Download the dataset manually from Kaggle.  
  After downloading, rename the file to `fraud_detect.csv` (if necessary) and place the file inside the `data/` folder at the root of the project.  

## Important
- Ensure that filenames match exactly; otherwise, notebooks may fail to run.  
- Ensure all downloaded files are properly placed inside the `data/` folder.  
- If you encounter any issues, refer back to this notebook for troubleshooting the datasets.
